```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
X = iris.data
labels_true = iris.target

kmeans = KMeans(n_clusters=3, random_state=42)
dbscan = DBSCAN(eps=0.6, min_samples=11)

kmeans_labels = kmeans.fit_predict(X)
dbscan_labels = dbscan.fit_predict(X)

print("K-Means聚类结果:", np.unique(kmeans_labels, return_counts=True))
print("DBSCAN聚类结果:", np.unique(dbscan_labels, return_counts=True))
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].scatter(X[:, 0], X[:, 1], c=labels_true, cmap='viridis')
axes[0].set_title('original')
axes[1].scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis')
axes[1].set_title('K-Means')
axes[2].scatter(X[:, 0], X[:, 1], c=dbscan_labels, cmap='viridis')
axes[2].set_title('DBSCAN')
plt.tight_layout()
plt.show()

kmeans_silhouette = silhouette_score(X, kmeans_labels)
dbscan_silhouette = silhouette_score(X, dbscan_labels)
print(f"\n轮廓系数评估:")
print(f"K-Means轮廓系数: {kmeans_silhouette:.4f}")
print(f"DBSCAN轮廓系数: {dbscan_silhouette:.4f}")
```

```python
from sklearn.preprocessing import StandardScaler

X_new = X[:, 2:]
scaler=StandardScaler()
X_scaled = scaler.fit_transform(X_new)
dbscan = DBSCAN(eps=0.5, min_samples=8)
dbscan_labels = dbscan.fit_predict(X_scaled)
ig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels_true, cmap='viridis', alpha=0.7)
scatter = axes[1].scatter(X_scaled[:, 0], X_scaled[:, 1], c=dbscan_labels, cmap='viridis', alpha=0.7)
plt.tight_layout()
plt.show()

dbscan_silhouette = silhouette_score(X_scaled, dbscan_labels)
print(f"DBSCAN轮廓系数: {dbscan_silhouette:.4f}")
```
